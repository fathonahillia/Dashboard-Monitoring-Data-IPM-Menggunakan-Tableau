```python
# add new column with certain value
df1= pd.read_excel('Harapan Lama Sekolah.xlsx')
df1['Komponen'] = 'Harapan Lama Sekolah'

df2 = pd.read_excel('Pengeluaran per Kapita Disesuaikan.xlsx')
df2['Komponen'] = 'Pengeluaran per Kapita Disesuaikan'

df3 = pd.read_excel('Rata-rata Lama Sekolah.xlsx')
df3['Komponen'] = 'Rata-rata Lama Sekolah'

df4 = pd.read_excel('Umur Harapan Hidup Saat Lahir (UHH).xlsx')
df4['Komponen'] = 'Umur Harapan Hidup Saat Lahir'

df5 = pd.read_excel('Indeks Pembangunan Manusia.xlsx')
df5['Komponen'] = 'Indeks Pembangunan Manusia'

df6 = pd.read_excel('LatLong.xlsx')
```

```python
# Convert uppercase to title style


df1['Provinsi/Kabupaten/Kota'] = df1['Provinsi/Kabupaten/Kota'].str.title()
df2['Provinsi/Kabupaten/Kota'] = df2['Provinsi/Kabupaten/Kota'].str.title()
df3['Provinsi/Kabupaten/Kota'] = df3['Provinsi/Kabupaten/Kota'].str.title()
df4['Provinsi/Kabupaten/Kota'] = df4['Provinsi/Kabupaten/Kota'].str.title()
df5['Provinsi/Kabupaten/Kota'] = df5['Provinsi/Kabupaten/Kota'].str.title()

df6['Provinsi/Kabupaten/Kota'] = df6['Provinsi/Kabupaten/Kota'].str.replace('KABUPATEN ', '')
df6['Provinsi/Kabupaten/Kota'] = df6['Provinsi/Kabupaten/Kota'].str.title()

df1= pd.merge(df1, df6, on ='Provinsi/Kabupaten/Kota', how ="left")
df2= pd.merge(df2, df6, on ='Provinsi/Kabupaten/Kota', how ="left")
df3= pd.merge(df3, df6, on ='Provinsi/Kabupaten/Kota', how ="left")
df4= pd.merge(df4, df6, on ='Provinsi/Kabupaten/Kota', how ="left")
df5= pd.merge(df5, df6, on ='Provinsi/Kabupaten/Kota', how ="left")


# Concat all data
df_full = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df_full['Provinsi'] = df_full['Provinsi'].str.title()
```
